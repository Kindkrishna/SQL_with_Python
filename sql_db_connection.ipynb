### To insert multiple rows into your MySQL table at once, 
you should use the **executemany()** method instead of **execute()**.

This is much more efficient than running a single INSERT statement over and over again in a loop.

Here is how you can do it right in your Jupyter Notebook:

Why use this approach?
**%s Placeholder**: These act as secure placeholders for your data. The connector automatically handles adding the single quotes around strings (like 'Patna' or 'Priya'), which prevents the syntax errors you ran into earlier.

**my_cursor.rowcount**: This is a handy property that will output exactly how many rows were just added to your student table.

```python
# 1. Define the SQL query with placeholders (%s)
sql = "INSERT INTO student (rollno, name, marks, grade, city) VALUES (%s, %s, %s, %s, %s)"

# 2. Prepare a list of tuples containing your data
data = [
    (111, 'Siddharth', 90, 'A', 'Patna'),
    (112, 'Priya', 85, 'B', 'Ranchi'),
    (113, 'Amit', 72, 'C', 'Gaya'),
    (114, 'Neha', 95, 'A', 'Muzaffarpur')
]

# 3. Use executemany to insert all rows at once
my_cursor.executemany(sql, data)

# 4. Commit the transaction to save changes
my_db.commit()

print(f"{my_cursor.rowcount} records inserted successfully.")
```

```python
my_cursor.execute('select * from student')
for i in my_cursor:
    print(i)
```

To delete records from your table, you use the SQL DELETE FROM statement. Just like with inserting data, it is highly recommended to use placeholders (%s) to specify which record(s) you want to remove. This prevents accidental syntax errors and protects against SQL injection.

Here are the two ways you can handle deletions in your Jupyter Notebook:

1. Delete a Single Specific Record
If you want to delete one specific student (for example, the student with rollno = 110), use this code:

```python
# Define the delete query with a placeholder
sql = "DELETE FROM student WHERE rollno = %s"

# Provide the value as a tuple (note the comma at the end for a single value tuple)
data = (110,)

# Execute and commit
my_cursor.execute(sql, data)
my_db.commit()

print(f"{my_cursor.rowcount} record(s) deleted.")
```

```python
my_cursor.execute('select * from student')
for i in my_cursor:
    print(i)
```

2. Delete Multiple Specific Records
If you want to delete multiple specific rows at once (for example, removing students with roll numbers 111 and 112), you can pair your query with executemany():

```python
# Define the delete query
sql = "DELETE FROM student WHERE rollno = %s"

# List of tuples containing the roll numbers to delete
data = [
    (111,),
    (112,)
]

# Execute multi-delete and commit
my_cursor.executemany(sql, data)
my_db.commit()

print(f"{my_cursor.rowcount} record(s) deleted.")
```

```python
my_cursor.execute('select * from student')
for i in my_cursor:
    print(i)
```

To update existing records in your table, you use the SQL UPDATE statement. Just like with inserting and deleting, using %s placeholders is the best practice to safely pass your data.

Here is how you can update a record (for example, updating a student's marks and grade based on their roll number) in your Jupyter Notebook:

1. Update a Single Record

```python
# 1. Define the UPDATE query with placeholders
sql = "UPDATE student SET marks = %s, grade = %s WHERE rollno = %s"

# 2. Provide the new values and the matching condition (New Marks, New Grade, Roll No)
data = (95, 'A', 101)

# 3. Execute and commit the changes
my_cursor.execute(sql, data)
my_db.commit()

print(f"{my_cursor.rowcount} record(s) updated successfully.")
```

```python
my_cursor.execute('select * from student')
for i in my_cursor:
    print(i)
```

2. Update Multiple Records at Once
If you have a list of multiple students whose records need to be updated, you can pair the UPDATE statement with executemany():

```python
# Define the query
sql = "UPDATE student SET marks = %s, grade = %s WHERE rollno = %s"

# List of tuples: [(New Marks, New Grade, Roll No), ...]
data = [
    (88, 'B', 104),
    (45, 'A', 105),  # Giving Dhruv a massive grace-mark update!
    (90, 'A', 106)
]

# Execute batch update and commit
my_cursor.executemany(sql, data)
my_db.commit()

print(f"{my_cursor.rowcount} records updated successfully.")
```

```python
my_cursor.execute('select * from student')
for i in my_cursor:
    print(i)
```

### 2. Reading SQL Data into a Pandas DataFrame
To pull your student table data into Pandas, use create_engine from SQLAlchemy to establish the connection, and then use pd.read_sql():

```python
import pandas as pd
from sqlalchemy import create_engine

# 1. Create a database connection engine
# Format: mysql+mysqlconnector://username:password@host/database_name
engine = create_engine("mysql+mysqlconnector://root:abc11@localhost/college")

# 2. Write your SQL Query
query = "SELECT * FROM student"

# 3. Load the data into a Pandas DataFrame
df = pd.read_sql(query, con=engine)

# Display the DataFrame
df
```

### 3. Writing a Pandas DataFrame back to a SQL Table
If you have data in a Pandas DataFrame and want to save it as a new table (or append it to an existing table) in your MySQL database, you use .to_sql():

```python
# Sample DataFrame to send to MySQL
new_data = pd.DataFrame({
    'rollno': [115, 116],
    'name': ['Karan', 'Sneha'],
    'marks': [82, 91],
    'grade': ['B', 'A'],
    'city': ['Ranchi', 'Patna']
})

# Write the DataFrame to a SQL table named 'student'
# if_exists='append' will add the rows without deleting existing data
new_data.to_sql(name='student', con=engine, if_exists='append', index=False)

print("Data successfully written to MySQL!")
```
